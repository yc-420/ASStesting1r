```python
def outliers_iqr(ys): #using +-1.5x IQR method for defining outliers

    quartile_1, quartile_3 = np.percentile(ys, [25, 75])
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    return np.where((ys > upper_bound) | (ys < lower_bound))

pos = outliers_iqr(ABC.actual_productivity)
pos
```

```python
#6.0 Model requirement 
#Regression analysis
X_ABC = ABC.drop('actual_productivity', axis=1)
y_ABC = ABC['actual_productivity']

print(X_ABC.shape) #confirm x and y is separate 
print(y_ABC.shape)
print(X_ABC.columns)
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X_ABC, y_ABC,test_size=0.2, random_state=42)

print(Xtrain.shape) #957 training
print(Xtest.shape) #240 testing
```

```python
#baseline (dummyregressor)
from sklearn.dummy import DummyRegressor

baseline = DummyRegressor(strategy="mean") #Build baseline mode (predict mean)

baseline.fit(Xtrain, ytrain) #Training through y and x

y_pred_base = baseline.predict(Xtest) #use data(test) predict
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np

mae = mean_absolute_error(ytest, y_pred_base)
rmse = np.sqrt(mean_squared_error(ytest, y_pred_base))
r2 = r2_score(ytest, y_pred_base)

print("Baseline Results:")
print("MAE:", round(mae,4))
print("RMSE:", round(rmse,4))
print("R2:", round(r2,4))
```

```python
#1. model (Linear Regression)
from sklearn.linear_model import LinearRegression

linreg = LinearRegression() # Build model
linreg.fit(Xtrain, ytrain) # Train
y_pred_lin = linreg.predict(Xtest) # Predict and from 1197 record find related average 

# Evaluate (sklearn Scoring formula)
mae = mean_absolute_error(ytest, y_pred_lin)
rmse = np.sqrt(mean_squared_error(ytest, y_pred_lin))
r2 = r2_score(ytest, y_pred_lin)

print("Linear Regression Results:")
print("MAE:", round(mae,4))
print("RMSE:", round(rmse,4))
print("R2:", round(r2,4))
```

```python
#cross validation
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

linreg = LinearRegression()

cv_r2 = cross_val_score(linreg, Xtrain, ytrain, cv=5, scoring='r2')
cv_rmse = -cross_val_score(linreg, Xtrain, ytrain, cv=5, scoring='neg_root_mean_squared_error')

print("Linear Regression 5-Fold CV R2:", cv_r2)
print("Linear Regression Average CV R2:", round(cv_r2.mean(), 4))
print("Linear Regression Average CV RMSE:", round(cv_rmse.mean(), 4))
```

```python
#2. Ridge Regression
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0) #punish
ridge.fit(Xtrain, ytrain)
y_pred_ridge = ridge.predict(Xtest)

mae = mean_absolute_error(ytest, y_pred_ridge)
rmse = np.sqrt(mean_squared_error(ytest, y_pred_ridge))
r2 = r2_score(ytest, y_pred_ridge)

print("Ridge Regression Results:")
print("MAE:", round(mae,4))
print("RMSE:", round(rmse,4))
print("R2:", round(r2,4))
```

```python
#parameter tuning
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

param = {'alpha': [0.01, 0.1, 1, 10, 100]}

ridge = Ridge()
grid_ridge = GridSearchCV(ridge, param, cv=5, scoring='r2')
grid_ridge.fit(Xtrain, ytrain)

print("Best Parameter:", grid_ridge.best_params_)
print("Best Estimator:", grid_ridge.best_estimator_)
print("Best CV R2:", grid_ridge.best_score_)
```

```python
#3. model (Random Forest Regressor)
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42) #use 100 test from 42
rf.fit(Xtrain, ytrain) #Train
y_pred_rf = rf.predict(Xtest)

mae = mean_absolute_error(ytest, y_pred_rf)
rmse = np.sqrt(mean_squared_error(ytest, y_pred_rf))
r2 = r2_score(ytest, y_pred_rf)

print("Random Forest Results:")
print("MAE:", round(mae,4))
print("RMSE:", round(rmse,4))
print("R2:", round(r2,4))
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(rf, Xtrain, ytrain, cv=5, scoring="r2")

print("Random Forest 5-Fold CV R2:", cv_scores)
print("Average CV R2:", cv_scores.mean())

#5-Fold Cross Validation was used to evaluate model robustness.
```

```python
#parameter tuning
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

param_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

rf = RandomForestRegressor(random_state=42)

grid_rf = GridSearchCV(
    estimator=rf,
    param_grid=param_rf,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

grid_rf.fit(Xtrain, ytrain)

print("Best Random Forest Parameters:", grid_rf.best_params_)
print("Best Random Forest Estimator:", grid_rf.best_estimator_)
print("Best Random Forest CV R2:", round(grid_rf.best_score_, 4))
```

```python
best_rf = grid_rf.best_estimator_
y_pred_rf = best_rf.predict(Xtest)

mae = mean_absolute_error(ytest, y_pred_rf)
rmse = np.sqrt(mean_squared_error(ytest, y_pred_rf))
r2 = r2_score(ytest, y_pred_rf)

print("Tuned Random Forest Results:")
print("MAE:", round(mae, 4))
print("RMSE:", round(rmse, 4))
print("R2:", round(r2, 4))
```

```python
#5. model (Decision Tree Regressor)
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)
dt.fit(Xtrain, ytrain)

y_pred_dt = dt.predict(Xtest)

mae = mean_absolute_error(ytest, y_pred_dt)
rmse = np.sqrt(mean_squared_error(ytest, y_pred_dt))
r2 = r2_score(ytest, y_pred_dt)

print("Decision Tree Results:")
print("MAE:", round(mae,4))
print("RMSE:", round(rmse,4))
print("R2:", round(r2,4))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)

cv_r2 = cross_val_score(dt, Xtrain, ytrain, cv=5, scoring='r2') #cross validation
cv_rmse = -cross_val_score(dt, Xtrain, ytrain, cv=5, scoring='neg_root_mean_squared_error')

print("Decision Tree 5-Fold CV R2:", cv_r2)
print("Decision Tree Average CV R2:", round(cv_r2.mean(), 4))
print("Decision Tree Average CV RMSE:", round(cv_rmse.mean(), 4))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

param_dt = {
    'max_depth': [3, 5, 7, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

dt = DecisionTreeRegressor(random_state=42)

grid_dt = GridSearchCV(
    estimator=dt,
    param_grid=param_dt,
    cv=5,
    scoring='r2',
    n_jobs=-1
)

grid_dt.fit(Xtrain, ytrain)

print("Best Decision Tree Parameters:", grid_dt.best_params_)
print("Best Decision Tree Estimator:", grid_dt.best_estimator_)
print("Best Decision Tree CV R2:", round(grid_dt.best_score_, 4))
```

```python
best_dt = grid_dt.best_estimator_
y_pred_dt = best_dt.predict(Xtest)

mae = mean_absolute_error(ytest, y_pred_dt)
rmse = np.sqrt(mean_squared_error(ytest, y_pred_dt))
r2 = r2_score(ytest, y_pred_dt)

print("Tuned Decision Tree Results:")
print("MAE:", round(mae, 4))
print("RMSE:", round(rmse, 4))
print("R2:", round(r2, 4))
```
